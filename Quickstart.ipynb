Welcome to the quick start guide for Assignment 3. This notebook will guide you through importing the SQL file into a local database and connect to it via `pandas` and `sqlalchemy`.

# Importing the SQL file

First, you'll need to set up your database. Assuming you've already installed MySQL following the [official instructions](https://dev.mysql.com/doc/refman/8.0/en/installing.html), you'll need to create a database and import the SQL file. You can do this by running the following commands in your terminal:

```bash
mysql -u root -p # You'll be prompted for your password
mysql> CREATE DATABASE IF NOT EXISTS `assignment3`;
mysql> USE `assignment3`;
mysql> SOURCE StackOverflow2010.sql;
```

Note that this might take a while, as the SQL file is quite large. Once it's done, you can move on to the next step.

# Connecting to the database

Now that you have the database set up, you can connect to it using `pandas` and `sqlalchemy`. You'll need to install `sqlalchemy` first, which you can do by running the following command in your terminal:

```bash
pip install sqlalchemy mysql-connector-python
```

Once that's done, you can connect to the database using the following code:

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('mysql+mysqlconnector://root:password@localhost:3306/assignment3')
```

Note that you'll need to replace `password` with your actual password. Once you've done that, you can run the following code to test your connection:

```python
df = pd.read_sql_query('SELECT * FROM posts', engine)
df.head()
```
